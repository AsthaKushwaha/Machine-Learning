```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x= np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#Functional
plt.plot(x,y,'r-')
plt.xlabel('X-label')
plt.ylabel('Y-Label')
plt.title("Plot")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object oriented
fig = plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.8])  #[Left,Bottom,width,Height]
axes.plot(x,y)
axes.set_xlabel("X-Label")
axes.set_ylabel("Y-Label")
axes.set_title("Title")
```

```python
fig=plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])


axes1.plot(x,y,'r')
axes1.set_title("Larger")

axes2.plot(y,x,'b')
axes2.set_title("Smaller")
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#plt.tight_layout()
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
axes
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title("First")
axes[1].plot(y,x)
axes[1].set_title("Second")
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='XSquare')
ax.plot(x,x**3,label='XCube')

ax.legend(loc=10)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_picture.png',dpi=200)
```

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange',linewidth=3,alpha=0.3,linestyle='--',marker='o',markersize=10,
       markerfacecolor='red') #linewidth =lw  linestyle=ls
```

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green',linewidth=3,alpha=0.3,linestyle='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
